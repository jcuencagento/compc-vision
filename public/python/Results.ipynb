This are some of the functions done to retrieve solutions and conclusions from survey responses.

````python
import csv

# Open the CSV file for reading
csv_file = 'survey_responses.csv'
# Lists to store data from each column
id = []
image_name = []
category = []
selection = []
ts = []

try:
    with open(csv_file, mode='r', newline='') as file:
        reader = csv.reader(file, delimiter=',', quotechar='"')
        # Skip the header row
        next(reader, None)
        for row in reader:
            # Assuming the CSV has three columns, adjust as needed
            id.append(row[0])
            image_name.append(row[1])
            category.append(row[2])
            selection.append(row[3])
            ts.append(row[4])

except FileNotFoundError:
    print(f"File '{csv_file}' not found.")
except Exception as e:
    print(f"An error occurred: {e}")
````

````python
import matplotlib.pyplot as plt
# Sample data
categories = ["nature", "people", "objects", "memes", "architecture", "animals", "censurable"]
selections = ["aws", "google", "azure", "none"]
# Count the occurrences of each combination of category and selection
category_selection_count = {}
category_data = category
selection_data = selection
for category, selection in zip(category_data, selection_data):
    key = (category, selection)
    category_selection_count[key] = category_selection_count.get(key, 0) + 1

# Separate the counts for each selection into separate lists
selection1_counts = [category_selection_count.get((category, "aws"), 0) for category in categories]
selection2_counts = [category_selection_count.get((category, "google"), 0) for category in categories]
selection3_counts = [category_selection_count.get((category, "azure"), 0) for category in categories]
selection4_counts = [category_selection_count.get((category, "none"), 0) for category in categories]
# Create the bar chart
plt.figure(figsize=(12, 8))
bar_width = 0.2
index = range(len(categories))
plt.bar(index, selection1_counts, bar_width, label='AWS')
plt.bar([i + bar_width for i in index], selection2_counts, bar_width, label='Google')
plt.bar([i + 2 * bar_width for i in index], selection3_counts, bar_width, label='Microsoft')
plt.bar([i + 3 * bar_width for i in index], selection4_counts, bar_width, label='Ninguna')
plt.xlabel('Categories')
plt.ylabel('Count')
plt.title('Category Selection Counts')
plt.xticks([i + 1.5 * bar_width for i in index], categories)
plt.legend()
plt.tight_layout()
plt.show()
````

````python
# Sample data
categories = ["nature", "people", "objects", "memes", "architecture", "animals", "censurable", "uploaded"]
selections = ["aws", "google", "azure", "none"]
# Count the occurrences of each selection
selection_counts = {}
category_data = category
selection_data = selection
for category, selection in zip(category_data, selection_data):
    key = (category, selection)
    category_selection_count[key] = category_selection_count.get(key, 0) + 1
# Sum the counts for each selection
selection_counts = {}
for selection in selections:
    total_count = sum(category_selection_count.get((category, selection), 0) for category in categories)
    selection_counts[selection] = total_count
# Create the bar chart
plt.figure(figsize=(12, 8))
bar_width = 0.2
index = range(len(selections))
plt.bar(index, [selection_counts["aws"], selection_counts["google"], selection_counts["azure"], selection_counts["none"]], bar_width, label='All Categories')
plt.xlabel('Selections')
plt.ylabel('Count')
plt.title('Category Selection Counts (Sum of All Categories)')
plt.xticks(index, selections)
plt.legend()
plt.tight_layout()
plt.show()
````
